```python
import pandas as pd

# Read CSV file
df = pd.read_csv(r"C:\Users\AI_LAB\Downloads\social_media_dataset_with_null_duplicates_outliers.csv")

print(df.head())
print(df.tail())
print(df.info())
print(df.shape)
print(df.columns)
print(df.describe())
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
df["Platform"] = df["Platform"].fillna(df["Platform"].mode()[0])

df["Sentiment"] = df["Sentiment"].fillna(df["Sentiment"].mode()[0])

print(df)
```

```python
print(df.duplicated())
df.drop_duplicates(inplace=True)
print(df)
```

```python
Q1 = df['Likes'].quantile(0.25)
Q3 = df['Likes'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Likes'] < lower) | (df['Likes'] > upper)]

print(outliers)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['Likes']] = scaler.fit_transform(df[['Likes']])

print(df[['Likes']])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Platform"] = le.fit_transform(df["Platform"])
print(df)
```
